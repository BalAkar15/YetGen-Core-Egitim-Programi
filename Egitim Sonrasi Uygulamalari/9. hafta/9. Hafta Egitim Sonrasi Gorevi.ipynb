# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda():
    def __init__(self,kanal=1,tvDurum=True,ses = 20,tvList = ["TRT 1","STAR TV","SHOW TV","KANAL D","ATV","FOX TV"]):
        self.kanal = kanal
        self.tvDurum = tvDurum
        self.ses = ses
        self.tvList = tvList
    def tvAc(self):
        if self.tvDurum == False:
            print("TV Açılıyor...")
            self.tvDurum = True
            time.sleep(2)
            print("TV Açıldı")
        else: 
            print("TV Zaten Açık")
    def tvKapat(self):
        if self.tvDurum == True:
            print("TV kapatılıyor...")
            self.tvDurum = False
            time.sleep(2)
            print("TV Kapatıldı")
        else: 
            print("TV Zaten Kapalı")
    def sesAyari(self):
        if self.tvDurum == False:
            print("Önce TV'yi açınız.")
        else:
            print(f"Ses seviyeniz şu anda: {self.ses} Lütfen istediğiniz ses miktarını giriniz: ")
            sesAyar = int(input("Lütfen istediğiniz ses miktarını giriniz: "))
            self.ses = sesAyar
            print(f"Ses seviyeniz {self.ses} seviyesine getirildi.")
    def kanalEkle(self):
        if self.tvDurum == False:
            print("Önce TV'yi açınız.")
        else:
            print("lütfen eklemek istediğiniz kanalı giriniz: ")
            kanal = input("lütfen eklemek istediğiniz kanalı giriniz: ")
            self.tvList.append(kanal)
            print(f"{kanal} Eklendi")
    def kanalSayisi(self):
        if self.tvDurum == False:
            print("Önce TV'yi açınız.")
        else:
            print(f"Kanal listenizde {len(self.tvList)} tane kanal bulunmaktadır.")
    def rastgeleKanal(self):
        if self.tvDurum == False:
            print("Önce TV'yi açınız.")
        else:
            rastgeleKanal = random.randint(1,len(self.tvList)-1)
            self.kanal = rastgeleKanal
            print(f"Şu anda {self.tvList[self.kanal]} kanalındasınız.")
    def tvInfo(self):
        if self.tvDurum == False:
            print("Önce TV'yi açınız.")
        else:
            print(f"Bu TV'de {len(self.tvList)} tane kanal bulunmaktadır.")
            print("----KANAL LİSTESİ----")
            count = 1
            for i in self.tvList:
                print(f"{count} : {i}")
                count +=1
            print("---------------------")
    
print("""
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
!!Çıkmak için q'ya basın!!""")

kumanda = Kumanda()
    
while True:
    sayi = input("Lütfen bir seçim yapın: ")
    if sayi == "q":
        break
    elif sayi == "1":
        kumanda.tvAc()
    elif sayi == "2":
        kumanda.tvKapat()
    elif sayi == "3":
        kumanda.sesAyari()
    elif sayi == "4":
        kumanda.kanalEkle()
    elif sayi == "5":
        kumanda.kanalSayisi()
    elif sayi == "6":
        kumanda.rastgeleKanal()
    elif sayi == "7":
        kumanda.tvInfo()
        
        
```
