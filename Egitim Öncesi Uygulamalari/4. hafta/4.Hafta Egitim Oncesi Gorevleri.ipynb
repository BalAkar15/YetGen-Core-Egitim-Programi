# Problem 1
Sözlükleri kullanarak bir telefon rehberi yazın. Bu rehberde kullanıcıya kimin telefonunu görüntülemek istediğini sorun ve kullanıcının girdiği isme göre o kişinin telefon numarasını yazdırın. Proje sonunda elde edeceğiniz çıktı  şuna benzer olmalı:


Lütfen numarasını öğrenmek istediğiniz kişinin adını girin: Ahmet

Ahmet isimli kişinin numarası şu şekildedir: 0532 678 13 19

````python
# 1.Problemin Çözümünü Buraya Yazınız
telRehber = {"ali": "05443967432", "ahmet": "05463221456", "şenol": "05303760793" }
kisi = input("Lütfen numarasını öğrenmek istediğiniz kişinin adını girin:")

if kisi in telRehber:
    print(f"{kisi} isimli kişinin numarası şu şekildedir: {telRehber[kisi]}")
else:
    kisiEkle = input(f"{kisi} adlı kişi rehberimizde kayıtlı değildir. Eğer numarasını kaydetmek istiyorsanız e tuşuna, eğer istemiyorsanız h tuşuna basınız.")
    if kisiEkle == "e":
        yeniKisi = input(f"Lütfen {kisi} adlı kişinin numarasını ekleyiniz.")
        telRehber[kisi] = yeniKisi
        print(f"{kisi} adlı kişinin numarası {yeniKisi} olarak başarıyla eklenmiştir.")
    elif kisiEkle == "h":
        print(f"{kisi} adlı kişinin numarası rehberimizde kayıtlı değildir.")


````
