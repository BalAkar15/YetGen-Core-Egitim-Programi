# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle():
    def __init__(self,name,speed,mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
    def introVehicle(self):
        print(f"{self.name.capitalize()} isimli araç Vehicle sınıfının bir örneğidir. {self.speed} km hızında ve {self.mileage} kilometrededir.")
class Bus(Vehicle):
    def __init__(self,color,name,speed,mileage):
        super().__init__(name,speed,mileage)
        self.color = color
    def seatingCapacity(self):
        print("The seating capacity of a bus is 50 passengers")
    def introBus(self):
        print(f"{self.name.capitalize()} markalı otobüs; {self.speed} km hızında, {self.mileage} kilometrede ve {self.color} renktedir. ")

mercedes = Bus("beyaz","mercedes",150,220)
print(mercedes.seatingCapacity())
print(mercedes.color)
print(mercedes.mileage)
mercedes.introBus()


````
